```python
import pandas as pd
import re

MAX_AMOUNT = 99999999999999.99

import psycopg2
from psycopg2 import sql
 
conn = psycopg2.connect(
    host="postgres",  
    database="mydb",
    user="myuser",
    password="mypassword",
    port="5432"  
)
 
cur = conn.cursor()

insert_query = '''
    INSERT INTO cargo (id, company_name, company_id, amount, status, created_at, paid_at)
    VALUES (%s, %s, %s, %s, %s, %s, %s)
'''

def is_valid_amount(amount):
    try:
        amount = float(amount)
                if amount < 0 or amount > MAX_AMOUNT:
            return False
        return True
    except ValueError:
        return False

def has_empty_values(row):
    required_columns = ['id', 'name', 'company_id', 'amount', 'status', 'created_at']
    for col in required_columns:
        if pd.isna(row[col]) or row[col] in [None, '']:
            return True, col, row[col]  
    return False, None, None 



def is_valid_row(row):
    alphanumeric_regex = re.compile(r'^[\w\s]+$')

    if not alphanumeric_regex.match(row['name']):
        return False, 'name', row['name']
    
    if not alphanumeric_regex.match(row['company_id']):
        return False, 'company_id', row['company_id']
    
    if not alphanumeric_regex.match(row['status']):
        return False, 'status', row['status']
    
    try:
        pd.to_datetime(row['created_at'])
    except (ValueError, TypeError):
        return False, 'created_at', row['created_at']
    
    if pd.notna(row['paid_at']):
        try:
            pd.to_datetime(row['paid_at'])
        except (ValueError, TypeError):
            return False, 'paid_at', row['paid_at']
    if '0xFFFF' in row['name']:
        return False, 'name', row['name']
    return True, None, None

#df_filtered = df[~df['name'].str.contains('0xFFFF', na=False)]

for index, row in df.iterrows():
    try:
        has_empty, empty_col, empty_value = has_empty_values(row)
        if has_empty:
            print(f"Fila {index} tiene un valor vacío en la columna '{empty_col}' con valor '{empty_value}', se omite la inserción.")
            continue  

        amount = row['amount']
        if not is_valid_amount(amount):
            print(f"Fila {index} tiene un valor de amount inválido: {amount}, se omite la inserción.")
            continue

        is_valid, error_column, error_value = is_valid_row(row)
        if not is_valid:
            print(f"Fila {index} tiene un error en la columna '{error_column}' con valor '{error_value}', se omite la inserción.")
            continue 

        paid_at = None if pd.isna(row['paid_at']) else row['paid_at']

        
        cur.execute(insert_query, (
            row['id'], 
            row['name'], 
            row['company_id'], 
            amount, 
            row['status'], 
            row['created_at'], 
            paid_at 
        ))

    except Exception as e:
        print(f"Error al insertar la fila {index}: {e}")

conn.commit()

cur.close()
conn.close()
```

```python
 
conn = psycopg2.connect(
    host="postgres",  
    database="mydb",
    user="myuser",
    password="mypassword",
    port="5432"  
)
cursor = conn.cursor()

create_view_query = """
CREATE VIEW total_transactions_per_day AS
SELECT 
    company_name,
    DATE(created_at) AS transaction_date,
    SUM(amount) AS total_amount
FROM 
    charges
JOIN 
    companies ON charges.company_id = companies.company_id
WHERE 
    status = 'paid'
GROUP BY 
    company_name, DATE(created_at)
ORDER BY 
    transaction_date, company_name;
"""

try:
    cursor.execute(create_view_query)
    conn.commit() 
    print("Vista creada exitosamente")
except Exception as e:
    print(f"Error al crear la vista: {e}")
    conn.rollback()  

cursor.close()
conn.close()
```

```python
 
conn = psycopg2.connect(
    host="postgres",  
    database="mydb",
    user="myuser",
    password="mypassword",
    port="5432"  
)
cursor = conn.cursor()

query = "SELECT * FROM total_transactions_per_day;"
cursor.execute(query)

columns = [desc[0] for desc in cursor.description]
results = cursor.fetchall()
df = pd.DataFrame(results, columns=columns)

print(df)

cursor.close()
conn.close()
```
